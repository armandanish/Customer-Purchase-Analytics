```python
import pandas as pd
df = pd.read_csv('customer_dataset.csv.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df = df.rename(columns ={'time_spent_on_product_research(hours)' : 'time_spent_on_product_research'})
```

```python
df.isnull().sum()
```

```python
df['social_media_influence'] = df['social_media_influence'].transform(lambda x: x.fillna(x.median()))
```

```python
df['social_media_influence'].dtype
```

```python
df['social_media_influence'] = pd.to_numeric(df['social_media_influence'], errors='coerce')
```

```python
df['social_media_influence'].fillna(df['social_media_influence'].median(), inplace=True)
```

```python
df['social_media_influence'].unique()
```

```python
df['social_media_influence'] = df['social_media_influence'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df = df.rename(columns={'time_spent_on_product_research(hours)' : 'time_spent_on_product_research'})
```

```python
df.columns
```

```python
df['purchase_amount'] = (df['purchase_amount'].str.replace('$','', regex=False).astype(float))
```

```python
df['social_media_influence'] = df['social_media_influence'].replace(0,'unknown')
```

```python
df['engagement_with_ads'].fillna('unknown', inplace=True)
```

```python
df['high_value_customer'] = (df['purchase_amount'] > df['purchase_amount'].median()).astype(int)
```

```python
df['purchase_efficiency'] = (df['purchase_amount'] / (df['time_spent_on_product_research']+1))
```

```python
df[['purchase_amount', 'time_spent_on_product_research', 'high_value_customer', 'purchase_efficiency']].head()
```

```python
#The purchase amount distribution is right-skewed, indicating that most customers make small to medium purchases, while a few customers make very high purchases.

import matplotlib.pyplot as plt
plt.hist(df['purchase_amount'], bins=20)
plt.title('Distribution of purchase amount')
plt.xlabel('Purchase Amount')
plt.ylabel('Number of customers')
plt.show()
```

```python
#Customers with higher engagement with ads tend to have a higher average purchase amount.

df.groupby('engagement_with_ads')['purchase_amount'].mean()
```

```python
#Social media influence shows a noticeable effect on customer spending behavior.

df.groupby('social_media_influence')['purchase_amount'].mean()
```

```python
#Time spent on product research shows a moderate relationship with purchase amount, indicating informed customers may spend more.

df[['time_spent_on_product_research', 'purchase_amount']].corr()
```

```python
#A relatively small portion of customers contribute to a large share of total revenue.

df['high_value_customer'].value_counts(normalize=True)
```
